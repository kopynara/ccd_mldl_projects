## üß© K-ÌèâÍ∑† Íµ∞Ïßë Î∂ÑÏÑù (K-Means Clustering)

### üìå 1. Íµ∞Ïßë(Clustering)Ïù¥ÎûÄ?

* üìö **ÎπÑÏßÄÎèÑ ÌïôÏäµ**: Ï†ïÎãµ(Î†àÏù¥Î∏î) ÏóÜÏù¥ Îç∞Ïù¥ÌÑ∞Ïùò Ìå®ÌÑ¥¬∑Íµ¨Ï°∞ Ï∞æÍ∏∞
* üë• **Íµ∞Ïßë Î∂ÑÏÑù**: ÎπÑÏä∑Ìïú ÌäπÏÑ±ÏùÑ Í∞ÄÏßÑ Îç∞Ïù¥ÌÑ∞ÎÅºÎ¶¨ Î¨∂Îäî Í≤É (Í∑∏Î£πÌïë)

---

### üìå 2. K-ÌèâÍ∑†(K-Means)Ïù¥ÎûÄ?

* üèÜ Í∞ÄÏû• Ïù∏Í∏∞ ÎßéÏùÄ Íµ∞Ïßë ÏïåÍ≥†Î¶¨Ï¶ò
* üéØ **Î™©Ìëú**: ÎØ∏Î¶¨ Ï†ïÌïú KÍ∞úÏùò Í∑∏Î£πÏúºÎ°ú ÎÇòÎàÑÍ∏∞
* üìç ÌïµÏã¨: \*\*Îç∞Ïù¥ÌÑ∞ ‚Üî Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï§ëÏã¨Ï†ê(Centroid)\*\*Ïóê Ìï†Îãπ

---

### ‚öô K-ÌèâÍ∑† ÏûëÎèô ÏàúÏÑú

1Ô∏è‚É£ **KÍ∞í Ï†ïÌïòÍ∏∞** ‚Üí Î™á Í∞úÏùò Í∑∏Î£πÏúºÎ°ú ÎÇòÎàåÏßÄ Í≤∞Ï†ï
2Ô∏è‚É£ **Ï§ëÏã¨Ï†ê(Centroid) Ï¥àÍ∏∞Ìôî** ‚Üí ÎûúÎç§ÏúºÎ°ú KÍ∞ú ÏÑ†ÌÉù
3Ô∏è‚É£ **Îç∞Ïù¥ÌÑ∞ Ìï†Îãπ** ‚Üí Í∞Å Îç∞Ïù¥ÌÑ∞Í∞Ä Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï§ëÏã¨Ï†êÏóê Î∂ôÏùå
4Ô∏è‚É£ **Ï§ëÏã¨Ï†ê Ïû¨Í≥ÑÏÇ∞** ‚Üí Í∞Å Í∑∏Î£πÏùò ÌèâÍ∑† ÏúÑÏπòÎ°ú Ïù¥Îèô
5Ô∏è‚É£ **Î∞òÎ≥µ** ‚Üí Ï§ëÏã¨Ï†êÏù¥ Îçî Ïïà Î≥ÄÌï† ÎïåÍπåÏßÄ 3\~4 Î∞òÎ≥µ

---

### üìä ÌäπÏßï

* ‚úÖ **Ïà´ÏûêÌòï Îç∞Ïù¥ÌÑ∞**ÏóêÏÑúÎßå ÎèôÏûë (Í±∞Î¶¨ Í≥ÑÏÇ∞ ÌïÑÏöî)
* üìè **Ïú†ÌÅ¥Î¶¨Îìú Í±∞Î¶¨**Î°ú Ïú†ÏÇ¨ÎèÑ ÌåêÎã®
* ‚ö° ÏÜçÎèÑ Îπ†Î¶Ñ, Íµ¨ÌòÑ Í∞ÑÎã®
* ‚ö† Ïù¥ÏÉÅÏπò(Outlier)Ïóê ÎØºÍ∞ê, KÍ∞íÏùÑ Ïûò Í≥®ÎùºÏïº Ìï® (ÏóòÎ≥¥Ïö∞ Í∏∞Î≤ï ÏÇ¨Ïö©)

---

### üîÑ Î®∏Ïã†Îü¨Îãù ÌîÑÎ°úÏÑ∏Ïä§ÏóêÏÑú Íµ∞Ïßë ÏúÑÏπò

1. üì• **Îç∞Ïù¥ÌÑ∞ ÏàòÏßë**
2. üõ† **ÌååÏù¥ÌîÑÎùºÏù∏**

   * Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ (Ïä§ÏºÄÏùºÎßÅ, Í≤∞Ï∏°Í∞í Ï≤òÎ¶¨)
   * **Íµ∞Ïßë Î∂ÑÏÑù (KMeans Îì±)**
3. üìå **ÏòàÏ∏°(Íµ∞Ïßë Î†àÏù¥Î∏î Ìï†Îãπ)**
4. üìè **ÌèâÍ∞Ä** (Ïã§Î£®Ïó£ Í≥ÑÏàò, SSE, ÎèÑÎ©îÏù∏ Ìï¥ÏÑù)

---

| Íµ¨Î∂Ñ       | ÏßÄÎèÑ ÌïôÏäµ(Î∂ÑÎ•ò¬∑ÌöåÍ∑Ä)      | ÎπÑÏßÄÎèÑ ÌïôÏäµ(Íµ∞Ïßë)                   |
| -------- | ----------------- | ---------------------------- |
| üéØ Î™©Ï†Å    | ÏûÖÎ†• ‚Üí Ï†ïÎãµ(Î†àÏù¥Î∏î) ÏòàÏ∏°   | ÏûÖÎ†• ‚Üí ÎπÑÏä∑Ìïú Îç∞Ïù¥ÌÑ∞ÎÅºÎ¶¨ Î¨∂Í∏∞            |
| üß† ÌïôÏäµ Î∞©Î≤ï | Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞(y)Î•º Î≥¥Î©¥ÏÑú ÌïôÏäµ | Ï†ïÎãµ(y) ÏóÜÏù¥ Îç∞Ïù¥ÌÑ∞Ïùò Íµ¨Ï°∞Î•º ÌïôÏäµ         |
| üìå Í≤∞Í≥ºÎ¨º   | ÏòàÏ∏° Î™®Îç∏ (Î∂ÑÎ•òÍ∏∞, ÌöåÍ∑ÄÏãù)  | Íµ∞Ïßë Î™®Îç∏ (Íµ∞Ïßë Ï§ëÏã¨Ï†ê, Í∞Å Îç∞Ïù¥ÌÑ∞Ïùò ÏÜåÏÜç Íµ∞Ïßë) |

# üß© Íµ∞ÏßëÎ∂ÑÏÑù(Clustering) Ï¥àÍ∞ÑÎã® Ïù¥Ìï¥

1. **üéØ Î™©Ìëú**

   * ÎπÑÏä∑Ìïú Ïï†Îì§ÎÅºÎ¶¨ **ÏûêÎèôÏúºÎ°ú** Î™®ÏïÑÏ£ºÎäî Í∏∞Ïà†
   * ÎùºÎ≤®(Ï†ïÎãµ) ÏóÜÏù¥ Îç∞Ïù¥ÌÑ∞Ïùò Ïà®ÏùÄ Í∑∏Î£π Ï∞æÍ∏∞ ‚Üí **ÎπÑÏßÄÎèÑ ÌïôÏäµ**

2. **‚öôÔ∏è ÏõêÎ¶¨** (K-Means ÏòàÏãú)
-   1Ô∏è‚É£ **K Ï†ïÌïòÍ∏∞**: ÎßåÎì§ Í∑∏Î£π Í∞úÏàò(K) ÎØ∏Î¶¨ Ï†ïÌï®
-   2Ô∏è‚É£ **Ï§ëÏã¨Ï†ê Î∞∞Ïπò**: ÎûúÎç§ÏúºÎ°ú KÍ∞úÏùò Ï§ëÏã¨(Centroid) Ï∞çÍ∏∞
-   3Ô∏è‚É£ **Í±∞Î¶¨ Í≥ÑÏÇ∞**: Î™®Îì† Îç∞Ïù¥ÌÑ∞Í∞Ä Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï§ëÏã¨Ïóê Î∂ôÏùå
-   4Ô∏è‚É£ **Ï§ëÏã¨ Ïû¨Í≥ÑÏÇ∞**: Í∞Å Í∑∏Î£π ÌèâÍ∑† ÏúÑÏπòÎ°ú Ï§ëÏã¨ ÏóÖÎç∞Ïù¥Ìä∏
-   5Ô∏è‚É£ **Î∞òÎ≥µ**: Ï§ëÏã¨Ïù¥ Îçî Ïù¥ÏÉÅ Ïïà ÏõÄÏßÅÏùº ÎïåÍπåÏßÄ 3\~4 Î∞òÎ≥µ

3. **üìè ÌäπÏßï**

   * üìä **Ïà´ÏûêÌòï Îç∞Ïù¥ÌÑ∞** ÏúÑÏ£º (Í±∞Î¶¨ Í≥ÑÏÇ∞ ÌïÑÏàò)
                       - Ïà´ÏûêÎßå Ïì∞Îäî Ïù¥Ïú†: Íµ∞ÏßëÏùÄ Î≥¥ÌÜµ Í±∞Î¶¨Î°ú ÎπÑÏä∑Ìï®ÏùÑ ÌåêÎã®ÌïòÍ∏∞ ÎïåÎ¨∏Ïù¥Ïïº.    
   * üé® Ïä§ÏºÄÏùº Ï°∞Ï†ï(Ï†ïÍ∑úÌôî/ÌëúÏ§ÄÌôî) Ï§ëÏöî
                       - Ïä§ÏºÄÏùºÎßÅ ÌïÑÏàòÍ∏â: ÌÅ¨Í∏∞Í∞Ä ÌÅ∞ Î≥ÄÏàò(Ïòà: price)Í∞Ä Í±∞Î¶¨ Í≥ÑÏÇ∞ÏùÑ ÏïïÎèÑÌïòÏßÄ ÏïäÍ≤å. 
   * üîç kÍ∞íÏùÄ ÏóòÎ≥¥Ïö∞(Elbow)¬∑Ïã§Î£®Ïó£(Silhouette) Îì±ÏúºÎ°ú Ï∞æÏùå
                       - k ÏÑ†ÌÉù: ÏóòÎ≥¥Ïö∞(SSE Í∫æÏûÑ), Ïã§Î£®Ïó£(0~1 ÎÜíÏùÑÏàòÎ°ù Ï¢ãÏùå) Í∞ôÏù¥ Îëò Îã§ Ï∞∏Í≥†ÌïòÎ©¥ ÏïàÏ†ïÏ†Å. 

4. **üí° ÌôúÏö© ÏòàÏãú**

   * üõç Í≥†Í∞ù Íµ¨Îß§ Ìå®ÌÑ¥ Í∑∏Î£πÌôî ‚Üí ÎßàÏºÄÌåÖ ÌÉÄÍ≤üÌåÖ
   * üéµ ÏùåÏïÖ Ï∂îÏ≤ú(ÎπÑÏä∑Ìïú Ï∑®Ìñ• Í∑∏Î£π)
   * üè† Î∂ÄÎèôÏÇ∞ Í∞ÄÍ≤©/ÌäπÏßïÎ≥Ñ ÏßÄÏó≠ Íµ¨Î∂Ñ

## üéØ SSE(Í¥ÄÏÑ±, Sum of Squared Errors)ÎûÄ?

* **Í∞Å Îç∞Ïù¥ÌÑ∞Í∞Ä ÏÜçÌïú Íµ∞ÏßëÏùò Ï§ëÏã¨Ï†ê(centroid)ÍπåÏßÄÏùò Í±∞Î¶¨**Î•º Í≥ÑÏÇ∞ÌïòÍ≥†
  Í∑∏ Í±∞Î¶¨Ïùò **Ï†úÍ≥±**ÏùÑ **Î™®Îì† Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ Ìï©Ìïú Í∞í**Ïù¥Ïïº.
* ÌïúÎßàÎîîÎ°ú, **Íµ∞Ïßë ÎÇ¥Î∂ÄÍ∞Ä ÏñºÎßàÎÇò Î≠âÏ≥êÏûàÎäîÏßÄ**Î•º Ïà´ÏûêÎ°ú ÎÇòÌÉÄÎÇ∏ Í≤É.

---

### üì¶ ÎπÑÏú†

* ÎßàÌä∏ Ïπ¥Ìä∏Ïóê Í≥ºÏùº(Îç∞Ïù¥ÌÑ∞)ÏùÑ Îã¥ÏïòÎã§Í≥† Ìï¥Î≥¥Ïûê.
* Ïπ¥Ìä∏ Ï§ëÏïô(centroid)Ïóê Í≥ºÏùºÎì§Ïù¥ **ÎπΩÎπΩÌïòÍ≤å Î™®Ïó¨ÏûàÏúºÎ©¥** ‚Üí SSEÍ∞Ä ÏûëÏùå ‚úÖ
* Í≥ºÏùºÎì§Ïù¥ ÏÇ¨Î∞©Ïóê Ìù©Ïñ¥Ï†∏ ÏûàÏúºÎ©¥ ‚Üí SSEÍ∞Ä ÌÅº ‚ùå

---

### üìâ ÌäπÏßï

1. **SSEÎäî ÏûëÏùÑÏàòÎ°ù Ï¢ãÏùå** ‚Üí Íµ∞Ïßë ÎÇ¥Î∂ÄÍ∞Ä Ïûò Î™®ÏòÄÎã§Îäî Îúª.
2. **KÎ•º ÎäòÎ¶¨Î©¥ SSEÎäî Î¨¥Ï°∞Í±¥ ÏûëÏïÑÏßê**
   (Íµ∞ÏßëÏù¥ ÎßéÏïÑÏßàÏàòÎ°ù Í∞Å Í∑∏Î£πÏù¥ ÏûëÏïÑÏ†∏ÏÑú Í±∞Î¶¨Í∞Ä Ï§ÑÏñ¥Îì¶)
3. ÌïòÏßÄÎßå ÎÑàÎ¨¥ ÏûëÏïÑÏßà ÎïåÍπåÏßÄ ÎäòÎ¶¨Î©¥ **Í≥ºÏ†ÅÌï©**Ï≤òÎüº ÏùòÎØ∏ ÏóÜÏùå ‚Üí Í∑∏ÎûòÏÑú **ÏóòÎ≥¥Ïö∞ Í∏∞Î≤ï**ÏúºÎ°ú Ï†ÅÎãπÌïú k Ï∞æÏùå.

---

### üßÆ ÏàòÏãù ÎäêÎÇå

```
SSE = Œ£(Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ÏôÄ Íµ∞Ïßë Ï§ëÏã¨Ï†ê Í±∞Î¶¨¬≤)
```

* Í±∞Î¶¨ Í≥ÑÏÇ∞ÏùÄ Ï£ºÎ°ú **Ïú†ÌÅ¥Î¶¨Îìú Í±∞Î¶¨**(ÏßÅÏÑ†Í±∞Î¶¨) ÏÇ¨Ïö©.

---

üí° Ìïú Ï§Ñ ÏöîÏïΩ:

> SSE = "Í∞ôÏùÄ Í∑∏Î£πÏóê ÏûàÎäî Îç∞Ïù¥ÌÑ∞Îì§Ïù¥ ÏñºÎßàÎÇò ÏÑúÎ°ú Í∞ÄÍπåÏù¥ Î™®Ïó¨ ÏûàÎäîÏßÄ"Î•º ÎÇòÌÉÄÎÇ¥Îäî Ï†êÏàò.
> ÎÇÆÏùÑÏàòÎ°ù Í∑∏Î£πÏù¥ Ï¥òÏ¥òÌïòÍ≥†, ÎÜíÏúºÎ©¥ Ìù©Ïñ¥Ï†∏ ÏûàÎã§Îäî Îúª.

---


````python
# ================== [0) ÌïúÍ∏Ä Ìè∞Ìä∏(ÏÑ†ÌÉù)] ==================
import matplotlib.pyplot as plt, platform
if platform.system() == 'Linux':
    plt.rcParams['font.family'] = 'NanumGothic'   # Ïö∞Î∂ÑÌà¨: ÎÇòÎàîÍ≥†Îîï
elif platform.system() == 'Darwin':
    plt.rcParams['font.family'] = 'AppleGothic'   # Îß•
else:
    plt.rcParams['font.family'] = 'Malgun Gothic' # ÏúàÎèÑÏö∞
plt.rcParams['axes.unicode_minus'] = False        # ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ


# ================== [1) ÎùºÏù¥Î∏åÎü¨Î¶¨ & Îç∞Ïù¥ÌÑ∞] ==================
import seaborn as sns                      # ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞ Î°úÎìúÏö©
import pandas as pd                        # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
from sklearn.preprocessing import StandardScaler  # Ïä§ÏºÄÏùºÎßÅ
from sklearn.cluster import KMeans         # K-ÌèâÍ∑† Íµ∞Ïßë
import matplotlib.pyplot as plt

# 1-1) Îç∞Ïù¥ÌÑ∞ Î°úÎìú (ÏµúÏ¥à 1ÌöåÎäî Ïù∏ÌÑ∞ÎÑ∑ÏóêÏÑú ÎÇ¥Î†§Î∞õÏïÑ Ï∫êÏãúÌï®)
dia = sns.load_dataset('diamonds')

# 1-2) Íµ∞ÏßëÏóê Ïì∏ "Ïà´Ïûê" Ïª¨ÎüºÎßå ÎΩëÍ∏∞ (Í±∞Î¶¨ Í≥ÑÏÇ∞Ïù¥Îùº Ïà´ÏûêÎßå Í∞ÄÎä•)
num_cols = ['carat','depth','table','x','y','z','price']
X = dia[num_cols].copy()
X.head()


# ================== [2) Ï†ÑÏ≤òÎ¶¨: Ïä§ÏºÄÏùºÎßÅ] ==================
# Í∞Å ÌäπÏÑ±Ïùò Îã®ÏúÑ/ÌÅ¨Í∏∞Í∞Ä Îã§Î•¥Îãà, ÌëúÏ§ÄÌôî(ÌèâÍ∑†0, ÌëúÏ§ÄÌé∏Ï∞®1)Î°ú ÎßûÏ∂§
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   # ÌïôÏäµ + Î≥ÄÌôò(ÌõàÎ†®ÏÖã Í∏∞Ï§Ä)



# ================== [3) k Í∞í Í∞ê Ïû°Í∏∞: ÏóòÎ≥¥Ïö∞(Elbow)] ==================
# kÎ•º 2~10ÍπåÏßÄ ÎèåÎ†§Î≥¥Í≥†, SSE(Í¥ÄÏÑ±)Í∞Ä "Îöù Í∫æÏù¥Îäî ÏßÄÏ†ê"ÏùÑ Í≥†Î¶Ñ
ks = range(2, 11)
inertias = []                          # Í∞Å kÏóêÏÑúÏùò SSE Ï†ÄÏû• Î¶¨Ïä§Ìä∏

for k in ks:
    km = KMeans(n_clusters=k, n_init='auto', random_state=42)  # kÍ∞ú Íµ∞Ïßë
    km.fit(X_scaled)                  # ÌïôÏäµ(Ï§ëÏã¨Ï†ê Ï∞æÍ∏∞)
    inertias.append(km.inertia_)      # Í¥ÄÏÑ±(SSE) Í∏∞Î°ù

plt.figure(figsize=(7,4))
plt.plot(ks, inertias, marker='o')
plt.title('ÏóòÎ≥¥Ïö∞(Elbow) Í≥°ÏÑ†')
plt.xlabel('k(Íµ∞Ïßë Ïàò)'); plt.ylabel('Í¥ÄÏÑ±(SSE, ÏûëÏùÑÏàòÎ°ù Ï¢ãÏùå)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()



# ================== [4) Ïã§Î£®Ïó£ Ï†êÏàòÎ°ú Ìïú Î≤à Îçî Ï≤¥ÌÅ¨(ÏÑ†ÌÉù)] ==================
# Í∞íÏù¥ 1Ïóê Í∞ÄÍπåÏö∏ÏàòÎ°ù Íµ∞ÏßëÏù¥ Ïûò Î∂ÑÎ¶¨Îê®. ÎåÄÎûµÏ†ÅÏù∏ ÎπÑÍµêÏö©.
from sklearn.metrics import silhouette_score

sil_scores = []
for k in ks:
    km = KMeans(n_clusters=k, n_init='auto', random_state=42)
    labels = km.fit_predict(X_scaled)     # Íµ∞Ïßë ÎùºÎ≤® Î∞îÎ°ú ÎΩëÍ∏∞
    sil_scores.append(silhouette_score(X_scaled, labels))

best_k = ks[int(max(range(len(sil_scores)), key=lambda i: sil_scores[i]))]
print(f"Ïã§Î£®Ïó£ Í∏∞Ï§Ä ÏµúÍ≥† k ÌõÑÎ≥¥: {best_k} (Ï†êÏàò={max(sil_scores):.3f})")

plt.figure(figsize=(7,4))
plt.plot(ks, sil_scores, marker='o')
plt.title('kÎ≥Ñ Ïã§Î£®Ïó£ Ï†êÏàò (ÎÜíÏùÑÏàòÎ°ù Ï¢ãÏùå)')
plt.xlabel('k'); plt.ylabel('Ïã§Î£®Ïó£')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()


# ================== [5) ÏµúÏ¢Ö Íµ∞ÏßëÌôî(Fit) + ÎùºÎ≤® Ï∂îÍ∞Ä] ==================
# ÏóòÎ≥¥Ïö∞/Ïã§Î£®Ïó£ Î≥¥Í≥† Ï†ÅÎãπÌïú k ÏÑ†ÌÉù (Ïòà: ÏúÑÏóêÏÑú ÎΩëÌûå best_k ÏÇ¨Ïö©)
K = best_k  # ÏßÅÍ¥ÄÏúºÎ°ú Î∞îÍøîÎèÑ Îê®. Ïòà: K=4
kmeans = KMeans(n_clusters=K, n_init='auto', random_state=42)

labels = kmeans.fit_predict(X_scaled)  # ÌïôÏäµ + ÏòàÏ∏°(Íµ∞Ïßë ÎùºÎ≤®)
dia['cluster'] = labels                # ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Íµ∞Ïßë Î≤àÌò∏ Ïª¨Îüº Ï∂îÍ∞Ä

# Í∞Å Íµ∞ÏßëÏùò "Ï§ëÏã¨(ÏÑºÌä∏Î°úÏù¥Îìú)"Î•º ÏõêÎûò Ïä§ÏºÄÏùºÎ°ú Î≥µÏõêÌï¥ Ìï¥ÏÑùÌïòÍ∏∞
centers_scaled = kmeans.cluster_centers_      # ÌëúÏ§ÄÌôî Í∏∞Ï§ÄÏùò Ï§ëÏã¨ Ï¢åÌëú
centers = scaler.inverse_transform(centers_scaled)  # ÏõêÎûò Îã®ÏúÑÎ°ú ÎêòÎèåÎ¶¨Í∏∞
centers_df = pd.DataFrame(centers, columns=num_cols)
centers_df


# ================== [6) ÏãúÍ∞ÅÌôî 1: carat vs price ÏÇ∞Ï†êÎèÑ] ==================
# ÏÉâÍπîÎ°ú Íµ∞Ïßë Íµ¨Î∂Ñ. Íµ∞ÏßëÎ≥Ñ Î∂ÑÌè¨Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú ÌôïÏù∏.
plt.figure(figsize=(8,5))
plt.scatter(dia['carat'], dia['price'], c=dia['cluster'], cmap='tab10', alpha=0.3, s=10)
plt.title(f'KMeans Íµ∞Ïßë Í≤∞Í≥º (k={K}) ‚Äî carat vs price')
plt.xlabel('carat'); plt.ylabel('price')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()



# ================== [7) ÏãúÍ∞ÅÌôî 2: PCA 2Ï∞®Ïõê Ìà¨ÏòÅ] ==================
# 7Ï∞®Ïõê(Ïó¨Îü¨ Ïà´Ïûê Ïª¨Îüº)ÏùÑ 2DÎ°ú Ï§ÑÏó¨ ÏÉâÏúºÎ°ú Íµ∞Ïßë ÌôïÏù∏
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=dia['cluster'], cmap='tab10', alpha=0.35, s=10)
plt.title(f'PCA Ìà¨ÏòÅÏóêÏÑú Î≥∏ Íµ∞Ïßë (k={K})')
plt.xlabel('PCA 1'); plt.ylabel('PCA 2')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()




# ================== [8) Î≥¥ÎÑàÏä§: KMeansÍ∞Ä ÌïôÏäµÎêòÎäî Î™®Ïäµ(Ï§ëÏã¨ Ïù¥Îèô) ==================
# 2Ï∞®Ïõê(carat, price)Îßå Í≥®Îùº ‚ÄúÏ§ëÏã¨Ïù¥ Ïñ¥ÎñªÍ≤å Ïù¥ÎèôÌïòÎäîÏßÄ‚Äù Îã®Í≥ÑÎ≥ÑÎ°ú ÏãúÍ∞ÅÌôî
import numpy as np
np.random.seed(42)

# 8-1) Í∞ÄÎ≥çÍ≤å 2000Í∞úÎßå ÏÉòÌîåÎßÅ
idx = np.random.choice(len(dia), size=2000, replace=False)
XY = dia.loc[idx, ['carat','price']].to_numpy()

# 8-2) Ïä§ÏºÄÏùº ÎßûÏ∂îÍ∏∞ (Í±∞Î¶¨ Í≥µÏ†ï)
sc = StandardScaler()
XYs = sc.fit_transform(XY)

# 8-3) Ï¥àÍ∏∞ Ï§ëÏã¨ KÍ∞ú Î¨¥ÏûëÏúÑ Î∞∞Ïπò
K_demo = min(4, len(XYs))                     # ÏòàÏãú k=4
centroids = XYs[np.random.choice(len(XYs), K_demo, replace=False)].copy()

def assign_clusters(points, centers):
    # Ìè¨Ïù∏Ìä∏-ÏÑºÌÑ∞ Í±∞Î¶¨(Ï†úÍ≥±Í±∞Î¶¨) Í≥ÑÏÇ∞ ‚Üí Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÏÑºÌÑ∞ index
    d2 = ((points[:,None,:] - centers[None,:,:])**2).sum(axis=2)
    return d2.argmin(axis=1)

plt.figure(figsize=(12,8))
for t in range(5):                            # 5Î≤àÎßå Î∞òÎ≥µ(Îç∞Î™®)
    lab = assign_clusters(XYs, centroids)     # ÌòÑÏû¨ Ï§ëÏã¨Ïóê Î∞∞Ï†ï
    new_centroids = centroids.copy()
    for k in range(K_demo):                   # Í∞Å Íµ∞ÏßëÏùò ÌèâÍ∑†ÏúºÎ°ú Ï§ëÏã¨ ÏóÖÎç∞Ïù¥Ìä∏
        pts = XYs[lab==k]
        if len(pts)>0:
            new_centroids[k] = pts.mean(axis=0)

    # ÌëúÏ§ÄÌôî ‚Üí ÏõêÏä§ÏºÄÏùºÎ°ú ÎêòÎèåÎ†§ Í∑∏Î†§Ï£ºÍ∏∞
    XY_plot = sc.inverse_transform(XYs)
    C_old   = sc.inverse_transform(centroids)
    C_new   = sc.inverse_transform(new_centroids)

    plt.subplot(2,3,t+1)
    plt.scatter(XY_plot[:,0], XY_plot[:,1], c=lab, cmap='tab10', alpha=0.25, s=8)
    plt.scatter(C_old[:,0], C_old[:,1], c='k', marker='x', s=80, label='Ïù¥Ï†Ñ Ï§ëÏã¨')
    plt.scatter(C_new[:,0], C_new[:,1], c='r', marker='o', s=60, label='ÏóÖÎç∞Ïù¥Ìä∏ Ï§ëÏã¨')
    for k in range(K_demo):
        plt.plot([C_old[k,0], C_new[k,0]], [C_old[k,1], C_new[k,1]], 'k--', alpha=0.6)
    plt.title(f'KMeans Îã®Í≥Ñ {t+1}')
    plt.xlabel('carat'); plt.ylabel('price')
    if t==0: plt.legend(fontsize=8)
    plt.grid(True, linestyle='--', alpha=0.3)

    centroids = new_centroids                 # Îã§Ïùå ÎùºÏö¥ÎìúÎ°ú

plt.tight_layout(); plt.show()


````
