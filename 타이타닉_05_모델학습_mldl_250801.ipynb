```python
# --- 1. 필요한 라이브러리 및 데이터 불러오기 ---
import pandas as pd
from sklearn.linear_model import LogisticRegression

# 훈련용 데이터를 CSV 파일에서 불러옵니다.
X_train = pd.read_csv('타이타닉_X_train_mldl_250801.csv')
y_train = pd.read_csv('타이타닉_y_train_mldl_250801.csv')
```

```python
# --- 2. 모델 선택 및 생성 ---

# # - 로지스틱 회귀(Logistic Regression)는 '생존(1) or 사망(0)'과 같은 이진 분류 문제에 자주 사용되는 모델입니다.
# # - random_state=42: 모델을 만들 때 무작위성을 제어해, 다음에 다시 실행해도 똑같은 결과를 얻을 수 있도록 합니다.

model = LogisticRegression(random_state=42)
```

```python
# --- 3. 모델 학습 ---

# # - .fit() 함수: 모델을 훈련시키는 함수입니다.
# # - 훈련 문제(X_train)와 훈련 정답(y_train)을 넣어 모델을 학습시킵니다.
# .values.ravel()을 추가해 데이터를 모델이 원하는 형태로 변환합니다.

model.fit(X_train, y_train.values.ravel())

print("--- 모델 학습 완료 ---")
```
