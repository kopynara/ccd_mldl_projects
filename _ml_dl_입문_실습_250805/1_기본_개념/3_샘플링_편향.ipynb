샘플링 편향 (Sampling Bias)

샘플링 편향이란, 데이터를 수집할 때 특정 데이터가 지나치게 많거나 적게 포함되어 전체 데이터의 특징을 제대로 반영하지 못하는 현상을 말합니다.

비유: 우리나라 전체 국민의 선호도를 조사할 때, 20대 남성만을 대상으로 설문조사하는 것과 같습니다. 
이 조사 결과는 20대 남성의 의견만 반영할 뿐, 전체 국민의 의견을 대표할 수 없습니다.

머신러닝에서의 문제: 모델이 편향된 데이터로 훈련되면, 실제 데이터에 대해 예측할 때 정확도가 크게 떨어질 수 있습니다. 마치 편향된 시각으로 세상을 보는 사람과 같습니다.

train_test_split과 stratify
우리가 이전에 사용했던 train_test_split 함수에서 stratify=y 매개변수는 바로 이 샘플링 편향을 방지하는 역할을 합니다. 
타겟(y)의 비율을 훈련 세트와 테스트 세트에 동일하게 분배하여, 데이터가 한쪽으로 치우치지 않게 해줍니다.

**stratify=y**를 사용하면, 데이터가 무작위로 나뉘면서 생길 수 있는 편향을 막아 모델을 더 공정하게 평가할 수 있습니다.
예를 들어, 타겟 y에 '클래스 0'이 70%, '클래스 1'이 30%의 비율로 있었다면, 
stratify=y를 적용하여 훈련 세트와 테스트 세트 모두 이 비율(70:30)을 유지하도록 데이터를 분리합니다.

stratify의 역할은 원래 데이터에 있던 타겟(y)의 비율을 그대로 유지하면서 데이터를 나누는 것입니다.
따라서 5:5로 나뉘는 것이 아니라, 전체 데이터의 비율을 따라간다고 생각하시면 됩니다.
