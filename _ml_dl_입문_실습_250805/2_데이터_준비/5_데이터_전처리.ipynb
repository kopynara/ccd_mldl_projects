데이터 전처리


데이터 전처리란, 

모델이 더 잘 이해하고 분석할 수 있도록 원본 데이터를 정리하고 변환하는 과정을 말합니다. 

마치 요리하기 전에 재료를 손질하는 것과 같습니다.


데이터 전처리가 중요한 이유

모델 성능 향상: 데이터의 잡음(noise)을 제거하고 중요한 특징(feature)을 강조하여 모델의 성능을 높입니다.

안정적인 훈련: 데이터의 스케일을 통일하여 모델이 특정 특성에만 치우치지 않도록 방지합니다.

데이터 품질 향상: 누락된 값(결측치)이나 이상치(outlier)를 처리하여 데이터의 신뢰도를 높입니다.
